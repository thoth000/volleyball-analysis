```python
# install package
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = "original_03.png"
origin_img = cv2.imread(path)
```

```python
# define court size
height = 1800
width = 900
```

```python
# cv2 : BGR
# matplotlib : RGB
plot_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)
plt.imshow(plot_img)
```

```python
cp_origin = np.float32([[209,1521], [688,2756], [2140, 2465], [1092, 1461]])
cp_trans = np.float32([[0,0], [0, height], [width, height], [width, 0]])
```

```python
M = cv2.getPerspectiveTransform(cp_origin, cp_trans)
trans_img = cv2.warpPerspective(origin_img, M, (width, height))
```

```python
# cv2 : BGR
# matplotlib : RGB
plot_img = cv2.cvtColor(trans_img, cv2.COLOR_BGR2RGB)
plt.imshow(plot_img)
```
