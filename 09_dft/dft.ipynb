```python
# 既知ポイント数
POINT_NUM = 6

# 既知ポイントの画像座標
iCoords = [
    [132, 774],
    [1090, 797],
    [903, 293],
    [515, 293],
    [100, 648],
    [1110, 669]
]
# 既知ポイントの実空間座標
rCoords = [
    [0, 0, 0],
    [0, 900, 0],
    [1800, 900, 0],
    [1800, 0, 0],
    [0, 0, 140],
    [0, 900, 140]
]
```

```python
# 計測行列について
def uArray(i):
    u = iCoords[i][0]
    return [rCoords[i][0], rCoords[i][1], rCoords[i][2], 1, 0, 0, 0, 0, -u*rCoords[i][0], -u*rCoords[i][1], -u*rCoords[i][2]]

def vArray(i):
    v = iCoords[i][1]
    return [0, 0, 0, 0, rCoords[i][0], rCoords[i][1], rCoords[i][2], 1, -v*rCoords[i][0], -v*rCoords[i][1], -v*rCoords[i][2]]

# 計測行列X
X = []
for i in range(POINT_NUM):
    X.append(uArray(i))
    X.append(vArray(i))
```
