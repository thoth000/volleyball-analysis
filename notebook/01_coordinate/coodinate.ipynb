```python
# install package
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = "original.png"
origin_img = cv2.imread(path)
```

```python
# define court size
height = 1800
width = 900
```

```python
# cv2 : BGR
# matplotlib : RGB
plot_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)
plt.imshow(plot_img)
```

```python
#LT, LB, RB, RT
cp_origin = np.float32([[69,707], [320, 2977], [2199, 2739], [1239, 779]])
cp_trans = np.float32([[0,0], [0, height], [width, height], [width, 0]])
```

```python
M = cv2.getPerspectiveTransform(cp_origin, cp_trans)
trans_img = cv2.warpPerspective(origin_img, M, (width, height))
```

```python
# cv2 : BGR
# matplotlib : RGB
plot_img = cv2.cvtColor(trans_img, cv2.COLOR_BGR2RGB)
plt.imshow(plot_img)
```

```python
write_path = "trans.png"
cv2.imwrite(write_path, trans_img)
```
