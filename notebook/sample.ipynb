```python
# install package
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sympy

%matplotlib notebook
```

```python
t1 = [400, 300, 360]
t2 = [0, 300, 370]
```

```python
x1 = [400, 0]
y1 = [300, -300]
z1 = [360, 0]

x2 = [0, 400]
y2 = [300, -250]
z2 = [240, 0]
```

```python
#3D Plot
fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
ax.plot(x1, y1, z1, "o-", color="Blue")
ax.plot(x2, y2, z2, "o-", color="Red")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
